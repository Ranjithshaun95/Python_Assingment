#### 5.	Create two arrays of any dimension and perform vstack and hstack

```python
arr1 = np.array([[1,2,3,4,5],['Six','Seven','Eight',9,10]])
arr2 = np.array([[1,'Two',3,4,'Five'],[6,7,'Eight',9,10]])
# Performing vstack
print('VStack Result:',np.vstack((arr1,arr2)))
# Performing hstack
print('HStack Result:',np.hstack((arr1,arr2)))
```
