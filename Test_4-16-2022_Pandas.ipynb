```python
import pandas as pd

## df is dataframe
df = pd.read_csv('C:\\Users\\Ranjith\\DATASCIENCE\\CSV\\Auto.csv')
```

```python
## 1. Find the most expensive car company name
#highestprice = max(df['price'])
#print(highestprice)
df = df [['make']][df.price==df['price'].max()]
df
```

```python
## 2.Count total cars per company
```

```python
df['make'].value_counts()
```

```python
## 3Find each company’s highest price car
```

```python
car_brand = df.groupby('make')
Df = car_brand['make','price'].max()
Df
```

```python
## 4. Sort all cars by Price column
```

```python
Df = Df.sort_values(by=['price'], ascending=False)
Df
```

```python
### 5. Group by make, fuel-type and body-style to calculate the average of city-mpg and highwaympg (miles per gallon)
```

```python
mpg=df.groupby(['make','fuel_type','body_style'])['city_mpg','highway_mpg'].mean()
mpg
```
